```python
def trainingExplosion2(dname, res_dict):

    training_dicts =[]
    for model in [Net7]:
        for d1 in [15]:
            for d2 in [50]:
                for d3 in [20]:
                    for d4 in [20]:
                        for num_layers in [4]:
                            for epochs in [80]:
                                for K in [20]:
                                    for seed in range(100):
                                        training_dicts.append(

                                            {
                                            # model params
                                            'class': model,
                                            'd1': d1,
                                            'd2': d2,
                                            'd3': d3,
                                            'd4': d4,
                                            'dK': K,
                                            'num_layers': num_layers,
                                            # now training params
                                            'dataset_name': dname,
                                            'epochs': epochs,
                                            'batch_size': 8,
                                            'seed': seed,
                                            }
                                            )

    for td in training_dicts:
        basic_analysis(
        td,
        res_dict,
        False
        )
```
