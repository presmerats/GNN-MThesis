# 3. Training setup options

```python
dname='../datasets/precomputed-pytorchg-betweenness_old/ws_100_3_0_1_nb.pickle'
res_dict={'tables':{}, 'scatterplots':{}}
visualize=True
basic_training({
        # model params
        'class': Net1,
        'd1': 20,
        # now training params
        'dataset_name': dname,
        'epochs': 20,
        'batch_size': 32,
        'seed': None,
        },
        res_dict,
        visualize
        )
reporting(res_dict)
```

```python
def trainingExplosion2(dname, res_dict):

    training_dicts =[]
    for model in [Net1]:
        for d1 in [15]:
            for d2 in [50]:
                for d3 in [20]:
                    for d4 in [20]:
                        for num_layers in [2,4,8,10]:
                            for epochs in [2,20]:
                                training_dicts.append(

                                    {
                                    # model params
                                    'class': model,
                                    'd1': d1,
                                    'd2': d2,
                                    'd3': d3,
                                    'd4': d4,
                                    'num_layers': num_layers,
                                    # now training params
                                    'dataset_name': dname,
                                    'epochs': epochs,
                                    'batch_size': 8,
                                    'seed': 35,
                                    }
                                    )

    for td in training_dicts:
        basic_training(
        td,
        res_dict,
        False
        )
```

```python
dname='../datasets/precomputed-pytorchg-betweenness_old/ws_100_3_0_1_nb.pickle'
res_dict={'tables':{}, 'scatterplots':{}}
trainingExplosion2(dname,res_dict)
reporting(res_dict)
```
